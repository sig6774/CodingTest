### 복잡도 
- 알고리즘의 성능을 나타내는 척도 
1. 시간복잡도 : 입력에 대해 알고리즘이 얼마나 오래 걸리는지 의미(연산 횟수)
2. 공간복잡도 : 입력에 대해 알고리즘이 얼마나 많은 메모리를 차지하는지 의미(메모리 양) 

```python
# 시간복잡도를 표현하기 위해서는 빅오 표기법을 사용 
# 해당 예제는 데이터의 크기에 따라서 시간 복잡도를 설명 
array = [3, 5, 1, 2, 4]
summary = 0

for x in array:
  summary += x

print(summary)
# 시간 복잡도는 O(N)이라 표기 
```

```python
a = 5
b = 7
print(a+b)
# 한번의 더하기 연산만 수행했으니 시간복잡도 O(1)
```

```python
array = [3, 6, 1, 2, 5]

for i in array:
  for j in array:
    temp = i*j
    print(temp)
# 시간 복잡도는 O(N^2)
```

```python
# 시간과 메모리 측정 
import time 
start = time.time()
end = time.time()
print('수행시간 :', end - start)
```

```python
# 정렬 알고리즘 
start = time.time()
for i in range(len(array)):
  min = i
  for j in range(i+1, len(array)):
    if array[min] > array[j]:
    # ex) 0번째 값이 다음 값보다 크면 
      min = j
  array[i], array[min] = array[min], array[i]

end = time.time()

print(end - start)
print(array)
```

### Greedy Algorithm

```python
# 거스름돈 
# 500, 100, 50, 10원짜리 동전이 무한히 존재하며 거슬러줘야 할 돈이 N원일때 거슬러줘야 할 동전의 최소 개수를 구하라 단, N은 항상 10배수 
# key : 최소 동전 개수이므로 최대 금액인 500원을 최대한 많이 사용 
n = 1260 
count = 0

coin = [500, 100, 50, 10]

for c in coin:
  count += n // c
  # 각 coin으로 n을 나눴을때 얻는 몫 
  n %= c
  # n을 c로 나눴을때 나머지 
  print(n)
print(count)
'''
1. coin의 가장 큰 값으로 나눠주며 몫을 구함 몫이 횟수를 뜻함 
2. 남은 나머지도 똑같이 수행하며 0이 될때까지 수행 

그리디 알고리즘은 거스름돈 문제에서는 매우 효과적이며 직관적이며 가장 큰 단위가 항상 작은 단위의 배수이므로 해결 가능 
가장 큰 단위부터 작은 단위까지 차례대로 확인하여 거슬러 주는 작업 수행한다는 아이디어 
'''
```

```python
def sum_max(num, m, k, d):
  d.sort()
  max = d[num-1]
  second_max = d[num-2]

  result = 0

  while True:
    for i in range(k):
      if m == 0:
        break 
      result += max
      m -= 1
    if m == 0:
      break 
    result += second_max
    m -=1
  return result
```

#### 숫자 카드 게임 
- 여러 숫자 카드 중 가장 높은 숫자가 쓰인 카드 한장을 뽑는 게임
- key : 각 행마다 가장 작은 수를 찾은 뒤 그 수 중에서 가장 큰 수를 찾는 것 

```python
n, m = map(int, input().split())

result = 0
for i in range(n):
  data = list(map(int, input().split()))

  min_value = min(data)
  # 각 행의 최소값 저장 

  result = max(result, min_value)
  # 각 행의 최소값 중 가장 큰 값 도출 
print(result)
```

#### 1이 될 때까지 
- 어떤 수 N이 1이 될 때까지 다음의 두 과정 중 하나를 반복적으로 선택하여 수행하려고 한다. 단, 두 번째 연산은 N이 K로 나누어떨어질 때만 선택할 수 있다. 
  1. N에서 1을 뺀다 
  2. N을 K로 나눈다 
- key : K로 최대한 나누기를 많이 수행 

```python
n, k = map(int, input().split())
result = 0

while n >= k:
  while n % k != 0:
    n -= 1
    result +=1 
  n //= k
  result += 1 

while n > 1:
  n -= 1
  result += 1 

print(result)
```

```python
n, k = map(int, input().split())
result = 0
while True:
  target = (n // k) * k
  # n이 k로 나눠떨어지기 위해 필요한 값 
  result += (n-target)
  # 1을 빼는 횟수 
  n = target

  if n < k:
    break 
  result += 1 
  n //= k 

result += (n - 1)
print(result)
'''
1. n이 k로 나누어 떨어지기 위한 값을 찾을 수 있도록 몫을 활용 
2. n에 찾은 값을 빼고 그 값이 -1씩 빼는 count 
3. 뺀값에 k를 나눠 몫을 도출 
4. 위의 과정을 while을 통해 반복하면 횟수 도출 가능 
'''
```
