#### 상하좌우 
- n*n 크기의 공간에서 특정 위치로 이동하였을때 좌표를 출력 


```python
n = int(input())

x, y = 1, 1
plans = input().split()

# L, R, U, D에 따른 이동 방향 
dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]
move_types = ['L', 'R', 'U', 'D']

for plan in plans:
  for i in range(len(move_types)):
    if plan == move_types[i]:
      nx = x + dx[i]
      ny = y + dy[i]
  if nx < 1 or ny < 1 or nx > n or ny > n: # 핵심 
    # 값이 범위보다 크거나 작게 되면 다시 돌아감 
    continue
    # 다시 첫번째 for문으로 이동 
  
  x, y = nx, ny 
print(x, y)
# 이동방향을 x,y로 나눠 특정 문자가 들어왔을때 조건을 만족하면 이동
```

#### 시각 
- 00시 00분 00초 ~ N시 59분 59초까지의 모든 시각 중 3이 하나라도 포함되는 모든 경우의 수를 구하는 프로그램을 작성 

```python
# time = int(input())
time = 5

count = 0
for h in range(1,time+1, 1):
  str_h = str(h)
  if str_h in '3':
    count += 1
  for m in range(0, 60, 1):
    str_m = str(m)
    if str_m in '3':
      count += 1
    for s in range(0, 60, 1):
      str_s = str(s)
      if str_s in '3':
        count+= 1
print(count)
# 내가 생각한 코드인데 이건 틀림
```

- 나이트 이동 범위 
  - 수평으로 두 칸 이동한 뒤 수직으로 한칸 
  - 수직으로 두칸 이동뒤 수평으로 한칸 
- 나이트가 이동할 수 있는 경우의 수를 출력

```python
# 수평 두칸 이동 후 수직 한칸 : (x+2, y-1), (x+2, y+1), (x-2, y+1), (x-2, y-1)
# 수직 두칸 이동 후 수평 한칸 : (x-1, y+2), (x+1, y+2), (x-1, y-2), (x+1, y-2)

steps = [(2, -1), (2, 1), (-2, 1), (-2, -1), (-1, 2), (1, 2), (-1, -2), (1, -2)] 
# 나이트 이동 경로 

data = input()
row  = int(data[1])
column = int(ord(data[0])) - int(ord('a')) + 1
# ord() : 문자열을 정수로 바꿔주는 함수 

result = 0
for step in steps:
  n_r = row + step[0]
  n_c = column + step[1]
  
  if n_r >= 1 and n_r <= 8 and n_c >= 1 and n_c <= 8: # 핵심 
    # 체스판의 크기에 맞을때 
    result += 1

print(result)
# 모든 경우의 수를 for문을 통해 대입해보고 if문을 통해 특정 조건에 맞을때만 횟수를 count 
```

```python
# 재귀함수 (자기 자신을 다시 호출하는 함수)

def recursive_function():
  print('재귀함수 호출')
  recursive_function()

recursive_function()
# 계속 자기 자신을 불러오기 때문에 재귀함수라 하며 print문이 계속 출력되는 것을 볼 수 있다.
```

```python
def recursive_function(i):
  if i == 100:
    return 
  print(i, '번째 재귀함수에서', i+1, '번째 재귀 함수를 호출합니다.')
  recursive_function(i+1)
  print(i, '번째 재귀함수를 종료합니다.')
recursive_function(1)
# 재귀함수는 스택 자료구조를 이용하기 때문에 첫번째 수행문을 모두 끝낸 이후 가장 먼저 들어온 99부터 다시 두번째 수행문을 수행 
# 즉, 재귀함수는 스택을 이용하여 첫번째 수행문의 값을 모두 출력하고 가장 최근에 들어온 값을 토대로 두번째 수행문 진행 
```

```python
def factorial_iterative(n):
  result = 1
  for i in range(1, n+1):
    result *= i
  # 1~n까지 계속 곱하기 
  return result
# 반복적 구현 

def factorial_recursive(n):
  if n <= 1:
    return 1
  return n* factorial_recursive(n-1)
# 재귀적 구현 
# 재귀함수의 장점 : 코드를 간결하게 사용가능 
```

```python
print('반복 : ', factorial_iterative(5))
print('재귀 : ', factorial_recursive(5))
```
