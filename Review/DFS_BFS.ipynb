### DFS 
- 탐색 : 많은 양의 데이터 중 원하는 데이터를 찾는 과정 
- 대표적인 그래프 탐색 알고리즘에는 DFS와 BFS가 존재 


#### 스택 
- 먼저 들어온 데이터가 나중에 나가는 형식(선입후출) 자료구조 
- 삽입과 삭제로 구성


```python
stack = []

stack.append(5)
stack.append(2)
stack.append(3)
print(stack)

stack.pop()
# 가장 최근에 들어간 3이 가장 먼저 나옴 

stack.append(1)

print(stack)
print(stack[::-1])
# 먼저 빠져나올 순서를 도출 
```

#### 큐 
- 먼저 들어온 데이터가 먼저 나가는 형식(선입선출)의 자료구조 
- 삽입과 삭제로 구성 

```python
from collections import deque 

queue = deque()

queue.append(5)
queue.append(2)
queue.append(3)
print(queue)

queue.popleft()
# 가장 먼저 들어간 5가 삭제된 것을 볼 수 있음 

print(queue)

queue.append(1)

print(queue)
queue.reverse()
print(queue)
```

#### 재귀함수 
- 자기 자신을 다시 호출하는 함수 
- DFS를 실제로 구현하고자 할 때 자주 사용 

```python
def recursive_function():
    print('재귀함수를 호출합니다.')
    recursive_function()
recursive_function()
# 이렇게 작성하면 똑같은 print문이 계속 나오므로 재귀함수를 멈출 조건을 작성해야함
# 함수가 재귀적으로 출력되면 스택형태로 동작
# 스택 자료구조 안에 함수의 정보가 차례로 담겨서 출력되는 것이라 생각 
```

```python
def recursive_function(i):
    # 100번째 출력했을 때 종료되도록 종료 조건 명시 

    if i == 100:
        return 
    print(i, '번째 재귀함수에서', i+1, '번째 재귀함수를 호출')
    recursive_function(i+1)
    print(i, '번째 재귀함수를 종료')
```

```python
recursive_function(1)
# 출력결과를 보면 스택처럼 100번째까지 데이터가 쌓이고 가장 최근에 들어온 i가 99이므로 99부터 빠지는 것을 확인할 수 있음 
```

```python
def factorial_iterative(n):
    result = 1 
    for i in range(1, n+1):
        result *= i
    return result

def factorial_recursive(n):
    if n <= 1:
        return 1 
    return n * factorial_recursive(n-1)
# 재귀함수를 사용하면 코드의 수를 효과적으로 줄일 수 있음
```

```python
print('반복으로 구현 : ', factorial_iterative(5))
print('재귀적으로 구현 : ', factorial_recursive(5))
```

#### 최대공약수 계산(유클리드 호제법)
- 두 자연수에 대한 최대 공약수를 구하는 것 
- A,B 두 자연수에 대해 A를 B로 나눈 나머지 R 
- A,B의 최대공약수는 B와 R의 최대공약수와 같음 

```python
def gcd(a,b):
    if a % b == 0:
        # a가 b의 배수라면 b가 a,b의 최대공약수 
        return b
    else:
        return gcd(b, a%b)
```

```python
print(gcd(192, 162))
```

### DFS(Depth-First Search)
- DFS : 깊이 우선 탐색이라 부르며 그래프에서 깊은 부분을 우선적으로 탐색하는 알고리즘 
- 동작 방법
    - 탐색 시작 노드를 스택에 삽입하고 방문 처리 
    - 최상단 노드에 방문하지 않은 인접 노드가 하나라도 있으면 그 노드를 스택에 넣고 방문 처리. 만약 방문하지 않은 인접 노드가 없으면 스택에서 최상단 노드를 꺼냄 
    - 2번 과정을 수행할 수 없을 때까지 반복

```python
def dfs(graph, v, visited):
    visited[v] = True 

    print(v, end = ' ')

    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i, visited)
```

```python
graph = [
    [],
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]
```

```python
visited = [False] * 9 
dfs(graph, 1, visited)
```

### BFS(Breadth-First Search)
- 너비 우선 탐색이라고 부르며 그래프에서 가까운 노드부터 우선적으로 탐색하는 알고리즘 
- 큐 자료구조를 이용
- 동작 과정 
    - 탐색 시작 노드를 큐에 삽입하고 방문 처리 
    - 큐에서 노드를 꺼낸 후 해당 노드의 인접 노드 중에서 방문하지 않은 노드를 모두 큐에 삽입하고 방문 처리 
    - 2번 과정을 수행할 수 없을 때까지 반복 


```python
from collections import deque

def bfs(graph, start, visited):

    queue = deque([start])
    # 현재 노드 방문 처리 

    visited[start] = True 

    while queue:
        v = queue.popleft()
        print(v, end = ' ')
        # 방문한 노드 번호 출력 

        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True 
```

```python
graph = [
    [],
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]
```

```python
visited = [False] * 9
bfs(graph, 1, visited)
'''순서 
1. graph[1]에 있는 2,3,8 방문처리 
2. graph[2]에 있는 1,7방문 처리 
3. graph[3]에 있는 1,4,5 방문 처리 
4. graph[8]에 있는 1,7은 방문 완료 했으니 통과 
5. graph[1]은 이미 했으니 통과 
6. graph[7]에 있는 2,6,8에서 6을 안했으니 6을 방문처리 
7. 나머지는 중복 
'''
```

#### 음료수 얼려 먹기 
- N*M 크기의 얼음틀이며 구멍이 뚫려 있는 부분은 0, 칸막이가 존재하는 부분은 1로 표시 
- 구멍이 뚫려 있는 부분끼리 상, 하, 좌, 우로 붙어 있는 경우 서로 연결되어 있는 것 
- 총 아이스크림 개수를 구하는 프로그램 

```python
n, m = map(int, input().split())

graph = [] 
for i in range(n):
    graph.append(list(map(int, input())))
```

```python
def dfs(x,y):
    if x <= -1 or x >= n or y <= -1 or y >= m:
        return False 
        # 틀을 벗어나면 False 
    if graph[x][y] == 0:
        graph[x][y] = 1
        # 방문 처리 

        dfs(x-1, y)
        dfs(x, y-1)
        dfs(x+1, y)
        dfs(x, y+1)
        return True 
        # 인접 부분에 대해서도 True이면 전체적으로 True 
    return False 
```

```python
result = 0
for i in range(n):
    for j in range(m):
        if dfs(i,j) == True:
            result += 1 
print(result)
```

#### 미로 탈출(이건 잘 모르겠음)
- n*m 크기의 직사각형 형태의 미로에 갇혔으며 여러 마리 괴물이 있어 괴물을 피해 탈출해야함 
- 현재 위치는 (1,1)이며 미로의 출구는 (n,m)의 위치에 존재하며 한번에 한칸씩 이동 
- 괴물이 있는 부분은 0, 없는 부분은 1로 표시 
- 탈출할 수 있는 최소의 칸의 개수를 구하면 됨 

```python
from collections import deque 

n,m = map(int, input().split())
graph = []
for i in range(n):
    graph.append(list(map(int, input())))
```

```python
# 상, 하, 좌, 우
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def bfs(x,y):
    queue = deque()
    queue.append((x,y))

    while queue:
        x, y = queue.popleft()
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]

            if nx < 0 or ny < 0 or nx >= n or ny >= m:
                continue 
            if graph[nx][ny] == 0:
                continue 
            if graph[nx][ny] == 1:
                graph[nx][ny] = graph[x][y] + 1 
                queue.append((nx, ny))

    return graph[n-1][m-1]
print(bfs(0,0))
```
