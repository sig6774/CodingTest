### Greedy Algorithm
- 현재 상황에서 지금 당장 좋은 것만 고르는 방법 
- 정당성 분석이 중요 
    - 단순히 가장 좋아 보이는 것을 반복적으로 선택해도 최적의 해를 구할 수 있는지 검토 
- 일반적으로는 탐욕법은 최적의 해를 보장할 수 없을 때가 많음 

#### 동전 나누기

```python
n = 1260 
count = 0

coin_type = [500, 100, 50, 10]

result = 0
for i in coin_type:
    result += n // i
    # 몫을 도출하는 것이 동전 개수 
    n %= i
    # 남은 나머지를 계속 사용하며 줄여줌 
print(result)

'''
500원부터 넣어서 동전의 개수를 구하는 이유는 
큰 단위가 작은 단위의 배수이므로 작은 단위를 종합해 다른 해가 나올 수 없음 
시간 복잡도 : O(K)
'''
```

#### 1이 될 떄까지 


```python
n, k = map(int, input().split())
```

```python
result = 0
while True:
    if n == 1:
        break 
        
    if n % k == 0:
        n //= k 
        result += 1
    else:
        n -= 1
        result += 1
print(result)
# 내 생각
# key : 나누는 것이 빼는 것보다 횟수를 줄일 수 있으므로 최대한 나누기를 수행 
```

```python
while n >= k:
    # n이 k로 나누어 떨어지지 않는다면 n에서 1씩 빼기 
    while n % k != 0:
        n -= 1
        result += 1
    n //= k 
    result += 1
while n > 1:
    n -= 1 
    result += 1
print(result)
# 책의 정답 
```

```python
while True:
    
    target = (n // k) * k 
    # key : 몫에 k를 곱하면 나누어떨어지는 값이 도출
    result += (n- target)
    # 1을 빼는것을 한번에 할 수 있음 
    n = target 
    
    if n < k:
        break 
        
    result += 1
    n //= k
    
result += (n-1)
# 반복문에서 나온 마지막 수에 대해 1씩 빼기 
print(result)
```

#### 곱하기 혹은 더하기 


```python
'''
기준을 정해야하지 않을까? 어디부터는 *를 한다던지...
아니네 0이 있으면 0과 앞 혹은 뒤에 있는 것을 더하고 나머지는 모두 곱하기 
'''
```

```python
a = input()
```

```python
cost = int(a[0])
for i in range(1,len(a)):
    if a[i] != '0':
        cost *= int(a[i])
    else:
        cost += int(a[i])
print(cost)
# 내 생각 
```

```python
data = input()

result = int(data[0])

for i in range(1, len(data)):
    num = int(data[i])
    if num <= 1 or result <= 1:
        result += num 
    else:
        result *= num 
print(result)
# 두 수에 대해서 두 수 중에서 하나라도 1이하인 경우에는 더하고 2 이상인 경우 곱하면 정답 
# 책의 정답인데 이게 더 좋은 코드이고 오류가 발생하지 않음 
```

```python
people.sort()
print(people)
```

```python
result = 0
count = 0 

for i in people:
    count += 1 
    if count >= i:
        result += 1 
        count = 0
print(result)
```
