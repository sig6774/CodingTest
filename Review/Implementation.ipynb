### 구현(Implementation)
- 머리속에 있는 알고리즘을 소스코드로 바꾸는 과정 
- 구현 문제 : 풀이를 떠올리는 것은 쉽지만 소스코드를 옮기기 어려운 문제 

```python
# 동, 북, 서, 남 
dx = [0, -1, 0, 1]
dy = [1, 0, -1, 0]
# 시뮬레이션 및 완전 탐색 문제에서는 2차원 공간에서의 방향 벡터가 자주 활용 

x, y = 2, 2 
for i in range(4):
    nx = x + dx[i]
    ny = y + dy[i]
    print(nx, ny)
```

#### 상하좌우 


```python
n  = int(input())
plans = input().split()

# L, R, U, D
dx = [0 ,0 ,-1, 1]
dy = [-1, 1, 0, 0]
move_types = ['L', 'R', 'U', 'D']
```

```python
x, y = 1, 1
for i in range(len(plans)):
    for j in range(len(move_types)):
        if plans[i] == move_types[j]:
            
            nx = x + dx[j]
            ny = y + dy[j]
            # 계속 갱신 
            
    if nx > n or ny > n or nx < 1 or ny < 1:
        continue 
    # 예외가 발생하면 두번째 반복문에서 빠져나옴 
    
    x, y = nx, ny
    # 이동 
    
print(x, y)
    
```

```python
h = int(input())
```

```python
result = 0
for i in range(h+1):
    for j in range(60):
        for k in range(60):
            time = str(i) + str(j) + str(k)
            # 문자로 바꿔주면 if문의 in 사용가능 
            if '3' in time:
                result += 1
print(result)
# 완전 탐색 유형(가능한 경우의 수를 모두 검사해보는 것)
```

#### 왕실의 나이트 


```python
input_data = input()
row = int(input_data[1])
column = int(ord(input_data[0])) - int(ord('a')) + 1

steps = [(-2,-1), (-2,1), (-1,-2), (-1,2), (1,-2), (1,2), (2,1), (2,-1)]
```

```python
result = 0
for i in steps:
    
    next_row = row + i[0]
    next_column = column + i[1]
    
    if next_row >= 1 and next_row <= 8 and next_column >= 1 and next_column <= 8:
        result += 1 
print(result)
```

```python
n.sort()
number = ['1', '2', '3', '4', '5', '6', '7', '8', '9','0']
sort_char = [0] * len(n)
for i in range(len(n)):
    for j in range(len(number)):
        if n[i] == number[j]:
            sort_char[-i] = n[i]
        else:
            sort_char[i] = n[i]
# 이거 아님 
```
